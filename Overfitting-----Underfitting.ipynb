```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch , Hyperband
```

```python
df = pd.read_excel(r'C:/BEPEC Python Material/Machine Learning/Logistic Regression/credit_card.xlsx', engine='openpyxl')
df
```

```python
column_names = df.loc[0,:]
column_names = column_names.to_list()
#print(type(x))
print(column_names)
print(type(column_names))
```

```python
l1 = df.columns
l1 = l1.to_list()
print(l1)
print(type(l1))
```

```python
df.columns = column_names
df.head(3)
```

```python
df.rename(columns= {'PAY_0':'PAY_1'}, inplace=True)
df.drop([0], inplace=True)
df.head(2)
df.drop(columns = {'ID'}, inplace=True)
```

```python
column_names = df.columns
for column in column_names:
    df[column] = pd.to_numeric(df[column]) 
df.info()
```

```python
df.head(2)
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
print(X.shape,y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
print(X.shape, X.dtype)
```

```python
def build_model(hp):
    model = keras.Sequential()
    for i in range(hp.Int('num_layers', 2, 20)):
        model.add(layers.Dense(units=hp.Int('units_' + str(i),
                                            min_value=10,
                                            max_value=500,
                                            step=5),
                               activation='elu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.compile(
        optimizer=keras.optimizers.Adam(
            hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
        loss='binary_crossentropy',
        metrics=['accuracy'])
    return model
```

```python
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=3,
    directory='project',
    project_name='Credit card default params',
    overwrite = False)
```

```python
tuner.search_space_summary()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```python
tuner.search(X_train, y_train,
             epochs=60, # This epoch will be greater 100,150,200
             validation_data=(X_test, y_test))
```

```python
best_hps = tuner.results_summary(num_trials = 1)
best_hps
```

```python
best_hp = tuner.get_best_hyperparameters()[0]
best_hp
```

```python
model = tuner.hypermodel.build(best_hp)
# model bulding through the best hyperparameters
```

```python
model.fit(X_train, y_train, validation_split = 0.1 ,epochs=60, batch_size=25, verbose=2)
```

```python
# evaluate the model
loss, acc = model.evaluate(X_test, y_test, verbose=2)
print('Test Accuracy: %.2f' % (acc*100))
print('loss is : %.2f'%loss)
```
